```python
import sympy as sp
```

```python
g,g_Y,theta=sp.symbols('g,g_Y,theta', positive=True)
```

Standard Model Reduced Neutral Gauge Boson Squared Mass Matrix

```python
M=(1/(g**2+g_Y**2))*sp.Matrix([[g**2,-g*g_Y],[-g*g_Y,g_Y**2]])
M
```

```python
O0,MD=M.diagonalize()
```

Modal Matrix $\begin{pmatrix}
\cos\theta_W & \sin\theta_W \\
-\sin\theta_W & \cos\theta_W
\end{pmatrix}$

```python
S=sp.simplify(O0*sp.sqrt(sp.transpose(O0)*O0)**(-1))*sp.Matrix([[0,1],[-1,0]])
S
```

Reduced Neutral Gauge Boson Squared Mass Matrix in Biagonal Basis

```python
sp.simplify(sp.transpose(S)*M*S)
```
